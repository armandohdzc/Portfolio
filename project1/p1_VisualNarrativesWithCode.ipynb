# Comparing Foods Technical Report

### Dataset origin & description
The dataset originally came from Kaggle, called Healthy Eating Dataset. This dataset consists of many different foods from different origins, and even shows some of the nutritional values of these foods. The values that are shown are calories, protein in grams, fiber, carbs and fats. With this data I plan to give an overview of some of the United States favorite cuisines and find if there is a "better" cuisine. 



```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

sns.set_style('darkgrid')
sns.set_palette("RdBu")
df = pd.read_csv('healthy_eating_dataset.csv')
```

### Preprocessing/cleaning steps (with rationale)
With this data set, I came across an issue that many would've probably loved to have. I had **too** much data, because of this I was not able to make any effective visuals. To solve this problem, I decided to only focus on three types of cuisines. I did this by searching what the top three cuisines are in the United States, after reading some articles I came to the conclusion that I should focus my report on Chinese, Mexican, and Italian foods.


```python
filtered = df[df["cuisine"].isin(["Mexican", "Chinese", "Italian"])]
```


### Visualization/design choices (why these charts and designs)
#### Why the boxplots?
I chose these visuals because I wanted to compare the interquartile range of the different cuisines for different nutritional values, by excluding the outlier and getting rid of the whiskers I am able to demonstrate that these foods are not actually that different when it comes down to it. For example, I can show the protein, calories, and fiber distribution using bar graphs

```python
a = sns.boxplot(x = "cuisine", y = "protein_g", data = filtered,whis = [25,75], showfliers = False, hue = "cuisine")
plt.xticks(rotation = 45)
a.set_title("Protein Distribution Across Different Cuisine")
a.set(xlabel = "Type of Cuisine", ylabel = "Protein in Grams")
plt.ylim(0,80)
plt.show()
```

```python
b = sns.boxplot(x = "cuisine", y = "calories", data = filtered,whis = [25,75], showfliers = False, hue = "cuisine")
plt.xticks(rotation = 45)
b.set_title("Calories Distribution Across Different Cuisine")
b.set(xlabel = "Type of Cuisine", ylabel = "Calories")
plt.ylim(0,1000)
plt.show()
```

```python
c = sns.boxplot(x = "cuisine", y = "fiber_g", data = filtered,whis = [25,75], showfliers = False, hue = "cuisine")
plt.xticks(rotation = 45)
c.set_title("Fiber Distribution Across Different Cuisine")
c.set(xlabel = "Type of Cuisine", ylabel = "Fiber in Grams")
plt.ylim(0,35)
plt.show()
```
