# What Foods should you be eating?
We will be discussing what types of foods you should be consuming. We will do this by looking at some of the most popular cuisine in the United States. In specific we will be looking at Chinese, Mexican and Italian cuisine, by narrowing down our search to these three we can make an informed decision by comparing the interquartile ranges of their nutritional values. 

### What are we looking for?

We are looking for foods that are nutrient dense - meaning that they are high in the amount of nutrients relative to its calorie content, such as foods high in protein or fiber but low in calories. By seeing which foods has a better, more consistent ratio we can make a decision on if there is a healthiest American favorite cuisine.

#### Lets start by looking at the calories...
Upon first glace we are able to see that Mexican and Italian have very similar medians but the Italians cuisine have a slightly longer interquartile range (25-75 percentile). If we look at Chinese, we see that they have a lower median than its competition and that it has a lower range than the other two.
