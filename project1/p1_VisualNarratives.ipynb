# What Type of Workouts Should You Be Doing

When we think about fitness, some people dread the thought of having to put in the effort and exert yourself. This thought alone is enough to make some people avoid physicality and gyms altogether. However, what if you knew how to maximize the efficiency of your workouts, this way you can put in the minimal amount of strain on your body while getting the maximum results. This brings up the next question, what do we classify as low effort, high reward? We can get an idea of this by comparing time spent working out, heart rate while working out, and calories burned. At the end we will also see if maybe age is a bigger factor than the type of workout you are doing when it comes to burning calories

Our data shows gym members acrosss a variety of different workout types including HIIT, cardio, strength training and yoga. By looking at metrics like heart rate, session duration, and calories burned, we can break it down into visuals and get a clear picture of which workouts provide the best efficiency.

### Calories Burned vs Workout Type
Looking at the raw data we can see that when we compare the four different workout types that there isnt a huge noticable difference. This is expected though, we are looking at a broad overview that includes outliers that can bring averages up and down. However, from this alone we can make the conclusion that HIIT workouts burn more calories on average, even if it is only slightly. 

### Session Time vs Calories Burned per hour
Taking a closer look into our data, we have broken up the workout sessions into half hour increments. This new view lets us know if workout out for longer periods of time proves to be more effective when it comes to burning calories per hour or if working out for shorter durations has the same effect on your body. From this visual we can see that there are no noticable jumps in calories burned per hours until we look at the 1.5-2 hour interval for HIIT workouts. HIIT workout however, go from being the least efficient workout timewise to being the most efficient once you get to longer periods of time. Which is great to find out, but we are looking for highest efficiency early on, and it seems like HIIT workouts become more efficient the longer you do them. From this visual we also see that early on cardio is taking a slight lead in the half hour to an hour range and then surpassed by strength training in the following time frames.

### Calories Burned vs Average Heart Rate
The purpose this visual is to represent exertion with heart rate, the higher your heart rate is, we can assume that you are working harder on your workout. This visual does bring up a concern however, there are no noticable consistent trends when it comes to calories burned and average heart rate. The only slightly noticable leader in these lines is the blue cardio one, we can see that it pokes up towards the top of the visual more consistently than the other types of workouts. In this visual we are measuring efficiency by how high the line is on the y axis, the higher a point is in the y, the more efficient this workout is per heart rate, meaning that in that heart rate range you burn more calories. 

### Calories Burned vs Age
Although these visuals have not been able to give us any conclusive evidence of any of them being significantly more efficient, we should consider another factor. This next and final factor is age, maybe being younger or older gives you a bigger advantage when it comes to burning calories? After breaking up the age ranges into 4 year intervals form 18 to 62, we can see that the younger somebody is they typically burn more calories than those that are older. This brings up a new question, is age the biggest factor when it comes to burning calories and seeing results in the gym? There's a chance that somebody could spend all the time they want doing what they think is the most efficient workout, and they will still see less results than somebody who is younger at the gym.

# Conclusion
In conclusion, we were unable to find a definitive "better" type of workout, we did see slight high points when comparing time to calories burned. However, the results were not drastic enough to make somebody change their workout routine. We did see that age may be the biggest factor when it comes to burning calories at the gym, the charts showed the biggest difference in data when we looked at the age versus calories burned.
